```python
import matplotlib.pyplot as plt
from sklearn import datasets
import numpy as np
```

```python
X = np.arange(0,100,5)
Y = np.arange(100,200,5)
```

```python
plt.scatter(X, Y, c=Y, cmap="RdYlBu")
```

```python
n_samples = 10000

x, y = datasets.make_circles(n_samples, noise=0.1, random_state=12)
```

```python
plt.scatter(x=x[:, 0], y=x[:,1], c=y, cmap=plt.cm.RdYlBu)
```

```python
NUM_CLASSES = 4
NUM_FEATURES = 2

X_blob, y_blob = datasets.make_blobs(n_samples=2000,
                                    n_features=NUM_FEATURES,
                                    centers=NUM_CLASSES,
                                    cluster_std=1.2,
                                     random_state=12
                                    )
```

```python
plt.figure(figsize=(12,8))
plt.scatter(X_blob[:,0],X_blob[:,1],c=y_blob,cmap=plt.cm.RdYlBu)
```
